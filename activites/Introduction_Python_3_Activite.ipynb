```python
# 1. Fonction affichant les nombres impairs ou pairs entre 1 et n
def n_print(n, odd_or_even):
    """
    Affiche tous les nombres impairs ou pairs entre 1 et n.

    Parameters:
    n (int): La limite supérieure de la plage de nombres à considérer.
    odd_or_even (str): Une chaîne indiquant si la fonction doit afficher des nombres "impair" ou "pair".
    """
    if odd_or_even == "impair":
        for i in range(1, n + 1, 2):
            print(i)
    elif odd_or_even == "pair":
        for i in range(2, n + 1, 2):
            print(i)

def n_print_while(n, odd_or_even):
    """
    Version alternative de n_print utilisant une boucle while.
    """
    i = 1
    while i <= n:
        if odd_or_even == "impair" and i % 2 != 0:
            print(i)
        elif odd_or_even == "pair" and i % 2 == 0:
            print(i)
        i = i + 1

# 2. Boucle for affichant le plus grand nombre dans une liste sans utiliser min() ou max()
def max_perso(mylist):
    """
    Retourne le plus grand élément d'une liste donnée.

    Parameters:
    mylist (list): Liste d'éléments numériques.

    Returns:
    int or float: Le plus grand élément de la liste.
    """
    maxMylist = mylist[0]
    for i in mylist:
        if i > maxMylist:
            maxMylist = i
    return maxMylist

# 3. Fonction retournant la somme de tous les nombres pairs dans une liste
def sum_even(lst):
    """
    Calcule et retourne la somme de tous les nombres pairs dans une liste donnée.

    Parameters:
    lst (list): Liste d'éléments numériques.

    Returns:
    int or float: Somme de tous les nombres pairs dans la liste.
    """
    i = 0
    sum_lst = 0
    while i < len(lst):
        if lst[i] % 2 == 0:
            sum_lst += lst[i]
        i += 1
    return sum_lst
```

#Dictionnary

```python
# 1. Écrivez une fonction Python qui prend un dictionnaire en entrée et
# retourne un nouveau dictionnaire où toutes les clés ont été converties en minuscules.
def lower_dict(input_dict):
    """
    Convertit toutes les clés d'un dictionnaire en minuscules.

    Parameters:
    input_dict (dict): Dictionnaire d'entrée avec des clés de type string.

    Returns:
    dict: Nouveau dictionnaire avec des clés en minuscules.
    """
    # Crée un dictionnaire vide pour stocker les paires clé-valeur en minuscules.
    lowercase_dict = {}
    # Itère sur les éléments du dictionnaire d'entrée.
    for key, value in input_dict.items():
        # Convertit la clé en minuscules et l'ajoute au nouveau dictionnaire avec la valeur originale.
        lowercase_dict[key.lower()] = value
    # Retourne le nouveau dictionnaire avec des clés en minuscules.
    return lowercase_dict

# 2. Écrivez une fonction Python qui prend un dictionnaire en entrée et
# retourne un nouveau dictionnaire où toutes les valeurs ont été doublées.
def double_dict(input_dict):
    """
    Double toutes les valeurs d'un dictionnaire.

    Parameters:
    input_dict (dict): Dictionnaire d'entrée avec des valeurs numériques.

    Returns:
    dict: Nouveau dictionnaire avec des valeurs doublées.
    """
    # Crée un nouveau dictionnaire pour stocker les valeurs doublées.
    doubled_values_dict = {}
    # Itère sur les éléments du dictionnaire d'entrée.
    for key, value in input_dict.items():
        # Double la valeur et l'ajoute au nouveau dictionnaire avec la même clé.
        doubled_values_dict[key] = value * 2
    # Retourne le nouveau dictionnaire avec des valeurs doublées.
    return doubled_values_dict

# Exemple d'utilisation pour doubler les valeurs dans un dictionnaire :
input_dict = {1: 1, 2: 2, 3: 3}
output_dict = double_dict(input_dict)
print(output_dict)  # Affichera {1: 2, 2: 4, 3: 6}
```
