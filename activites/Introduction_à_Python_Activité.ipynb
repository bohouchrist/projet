```python
# 1 Définissez une fonction appelée euro_dollars qui convertit des euros en dollars (Karl)
def euro_dollars(amount_euros):
"""
    Convertit un montant en euros en dollars.

    Parameters:
    amount_euros (float): Montant en euros à convertir.

    Returns:
    float: Montant converti en dollars.
    """
    # La fonction prend un montant en euros et le multiplie par le taux de change (1.06) pour le convertir en dollars
    final_amount = amount_euros * 1.06
    return final_amount

# 2 Essayez-le pour euro_dollars(1), euro_dollars(3.5), euro_dollars(10)
# Appel de la fonction euro_dollars avec différents montants en euros et affichage des résultats
print(euro_dollars(amount_euros=1))
print(euro_dollars(3.5))
print(euro_dollars(amount_euros=10))

# 3 Définissez une fonction appelée dollars_euro qui convertit des dollars en euros (Hoda)
def dollars_euro(dollar):
"""
    Convertit un montant en dollars en euros.

    Parameters:
    dollar (float): Montant en dollars à convertir.

    Returns:
    float: Montant converti en euros.
    """
    # La fonction prend un montant en dollars et le multiplie par le taux de change (0.84) pour le convertir en euros
    return dollar * 0.84

# 4 Essayez-le pour dollars_euro(1), dollars_euro(3.5), dollars_euro(10) (Galaad)
# Appel de la fonction dollars_euro avec différents montants en dollars et affichage des résultats
print(dollars_euro(1))
print(dollars_euro(3.5))
print(dollars_euro(10))

# 5 Passez d'une fonction à l'autre en fonction de la devise que vous souhaitez convertir
def convert_euro_dollar(amount, is_dollar=True):
"""
    Convertit un montant en euros ou en dollars en fonction du paramètre is_dollar.

    Parameters:
    amount (float): Montant à convertir.
    is_dollar (bool): Si True, convertit en euros, sinon en dollars.

    Returns:
    float: Montant converti.
    """
    # Si is_dollar est True, appelez la fonction dollars_euro, sinon appelez la fonction euro_dollars
    if is_dollar == True:
        res = dollars_euro(amount)
    else:
        res = euro_dollars(amount)
    return res

# 6 Convertissez des montants en euros ou en dollars en fonction du signe monétaire (Yasmine)
def convert_dollars_euros(amount, sign):
"""
    Convertit un montant en euros ou en dollars en fonction du signe fourni.

    Parameters:
    amount (float): Montant à convertir.
    sign (str): Signe de la devise cible ("€" ou "$").

    Returns:
    str: Montant converti avec le signe de la devise.
    """
    # Si le signe est "€", convertissez le montant en dollars, sinon convertissez-le en euros
    if sign == "€":
        convert_amount = str(amount * 1.06) + "$"
    else:
        convert_amount = str(amount * 0.94) + "€"
    return convert_amount

# 7 Convertissez une devise en une autre (Hoda)
def convert_currency(amount, currency_from, currency_to):
"""
    Convertit un montant d'une devise à une autre.

    Parameters:
    amount (float): Montant à convertir.
    currency_from (str): Devise d'origine ("euro" ou "dollar").
    currency_to (str): Devise cible ("euro" ou "dollar").

    Returns:
    float or str: Montant converti ou message d'erreur si la conversion est invalide.
    """
    # Si la devise source est "euro" et la devise cible est "dollar", appelez euro_dollars
    if currency_from == "euro" and currency_to == "dollar":
        return euro_dollars(amount)
    # Si la devise source est "dollar" et la devise cible est "euro", appelez dollars_euro
    elif currency_from == "dollar" and currency_to == "euro":
        return dollars_euro(amount)
    else:
        return "Conversion invalide"

# 8 Convertissez un montant en euros ou en dollars en fonction du signe monétaire (Amir)
# En fonction de la devise (euros ou dollars), convertissez le montant et ajoutez le signe monétaire correspondant
def amount(amount_euros, amount_dollars, currency):
"""
    Convertit un montant en euros ou en dollars en fonction du paramètre currency.

    Parameters:
    amount_euros (float): Montant en euros.
    amount_dollars (float): Montant en dollars.
    currency (str): Devise cible ("euros" ou "dollars").

    Returns:
    str: Montant converti avec le nom de la devise.
    """
    if currency == "euros":
        final_amount = amount_euros * 1.06
        return str(final_amount) + " dollars"
    else:
        final_amount = amount_dollars * 0.84
        return str(final_amount) + " euros"

# 9 Convertissez des montants en euros ou en dollars en fonction de la devise (Axel)
# Les fonctions euro_dollars et dollars_euro ont été répétées, donc nous ne les répétons pas ici
# La fonction convert_currency appelle soit euro_dollars, soit dollars_euro en fonction de la devise
def convert_currency(amount, currency):
"""
    Convertit un montant en euros ou en dollars en fonction du paramètre currency.

    Parameters:
    amount (float): Montant à convertir.
    currency (str): Devise cible ("euro" ou "dollar").

    Returns:
    float or str: Montant converti ou message d'erreur si la devise est incorrecte.
    """
    if currency == "euro":
        return euro_dollars(amount)
    elif currency == "dollar":
        return dollars_euro(amount)
    else:
        return "wrong devise"
```

# Liste
### Une liste est une collection ordonnée d'éléments, comme des chiffres ou des mots, regroupés entre crochets [ ]. Les listes sont flexibles et permettent de stocker différentes données. Chaque élément a sa propre position, appelée indice, dans la liste, ce qui facilite l'accès et la manipulation des informations.

```python
# Variables d'aire (en mètres carrés)
# Aire du hall
hall = 11.25
# Aire de la cuisine
kit = 18.0
# Aire du salon
liv = 20.0
# Aire de la chambre
bed = 10.75
# Aire de la salle de bains
bath = 9.50

# Créer une liste appelée "areas" contenant les aires des différentes pièces
areas = [hall, kit, liv, bed, bath]

# Afficher les aires en utilisant la fonction "print"
print(areas)

# Informations sur la maison sous forme de liste de listes
maison = [["couloir", hall],
         ["cuisine", kit],
         ["salle de séjour", liv]]

# Afficher la liste maison
print(maison)

# Afficher le type de la variable maison
print(type(maison))
```
