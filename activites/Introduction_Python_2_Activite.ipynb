`````python
# Cette fonction is_even() prend un nombre en entrée et renvoie True si le nombre est pair, sinon False.
# Elle vérifie si le reste de la division de 'num' par 2 est égal à zéro.
# Exemple d'utilisation :
# - is_even(2) renvoie True car 2 est pair.
# - is_even(3) renvoie False car 3 est impair.

def is_even(num):
    if num % 2 == 0:
        result = True
    else:
        result = False
    return result

# Exemple d'utilisation de la fonction is_even.
print(is_even(2))
# Affiche True
print(is_even(3))
 # Affiche False

# Une version améliorée de la fonction is_even() :
# Elle renvoie directement le résultat de la comparaison 'number % 2 == 0',
# ce qui simplifie le code en évitant une variable temporaire.
# Exemple d'utilisation :
# - is_even(4) renvoie True car 4 est pair.
# - is_even(5) renvoie False car 5 est impair.

def is_even(number):
    return number % 2 == 0

# Exemple d'utilisation de la version améliorée de la fonction is_even.
print(is_even(4))
 # Affiche True
print(is_even(5))
# Affiche False
`````

`````python
# Consigne : Fonctions pour Manipuler des Nombres

# 1. Écrivez une fonction avec deux arguments qui imprime tous les nombres pairs
# ou impairs entre 1 et le premier paramètre. Le deuxième paramètre vous indique
# si vous devez imprimer les nombres pairs ou impairs.

def n_print(n, parite):
    """
    Cette fonction imprime tous les nombres pairs ou impairs entre 1 et n
    en fonction du deuxième paramètre (parite).

    Args:
    n (int): Le nombre jusqu'auquel imprimer les nombres pairs ou impairs.
    parite (str): Indique si l'on doit imprimer les nombres "pair" ou "impair".

    Returns:
    None
    """
    if parite == "pair":
        for i in range(1, n):
            if i % 2 == 0:
                print(i)
    elif parite == "impair":
        for i in range(1, n):
            if i % 2 == 1:
                print(i)
    else:
        print("Saisie incorrecte")

# Exemple d'utilisation de la fonction n_print.
n_print(5, "pair")

# 2. Écrivez une boucle for qui imprime le plus grand nombre dans une liste
# sans utiliser les fonctions min() ou max().

myList = [4, 5, 3, 5, 2, 50]
def max_perso(mylist):
    """
    Cette fonction trouve et retourne le plus grand nombre dans une liste.

    Args:
    mylist (list): La liste de nombres dans laquelle chercher.

    Returns:
    int: Le plus grand nombre dans la liste.
    """
    maxMylist = mylist[0]
    for i in mylist:
        if i > maxMylist:
            maxMylist = i
    return maxMylist

# Exemple d'utilisation de la fonction max_perso.
max_perso([1, 2, 3])

# Utilisation de list comprehensions pour simplifier la recherche de nombres pairs et impairs.

def n_print(n, parite):
    """
    Cette fonction retourne une liste de nombres pairs ou impairs entre 1 et n
    en fonction du deuxième paramètre (parite).

    Args:
    n (int): Le nombre jusqu'auquel générer les nombres pairs ou impairs.
    parite (str): Indique si l'on doit générer les nombres "pair" ou "impair".

    Returns:
    list: Une liste de nombres pairs ou impairs.
    """
    if parite == "pair":
        return [i for i in range(2, n + 1, 2)]
    elif parite == "impair":
        return [i for i in range(1, n + 1, 2)]
    else:
        return "Saisie incorrecte"

# Exemple d'utilisation de la fonction n_print avec list comprehensions.
print(n_print(10, "impair"))

# Utilisation de list comprehensions pour filtrer les nombres pairs et impairs.

list_nb = [1, 2, 3, 4, 5, 6, 7]

even_numbers = [nb for nb in list_nb if nb % 2 == 0]
odd_numbers = [nb for nb in list_nb if nb % 2 != 0]

# Impression des nombres pairs et impairs.
print(even_numbers)
print(odd_numbers)
`````
